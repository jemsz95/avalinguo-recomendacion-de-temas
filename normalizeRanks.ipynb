```python
import json
import math
```

```python
with open('normalizedRanks.json', encoding="utf8", mode='r') as r:
    ranksFile= json.load(r)
```

```python
ranks= ranksFile
```

```python
for key in ranks:
    ranks[key]= math.log(ranks[key], 2)
```

```python
for key in ranks:
    print(ranks[key])
```

```python
with open('normalizedRanks.json', 'r+') as file:
    file.write(json.dumps(ranks))
    file.truncate()
```
